# N Queen Attack Detection

```python
import random
def generate_positions(N):
    r = []
    c = []
    for i in range(1, N+1):    
        r.append(random.choice(range(1, N+1)))
        c.append(random.choice(range(1, N+1)))
        #check repetition of position
        print('Q{}: {},{}'.format(i,r[-1],c[-1]))
    return r,c
```

```python
N = 4
row, col = generate_positions(N)
print(row)
print(col)
```

```python
# row attack
def rowattack(r, N):
    flag = False
    print('\nrow attack: (if any)')
    for i in range(0,N):
        for j in range(i+1,N):
            if row[i] == row[j]:
                  flag = True
                  print(f'row:{r[i]} column:{c[i]} and {c[j]}')
    return flag     
```

```python
#column attack
def columnattack(c, N):
    flag = False
    print('\ncolumn attack: (if any)')
    for i in range(0,N):
        for j in range(i+1,N):
            if(c[i] == c[j]):
                flag = True
                print(f'column:{c[i]} row:{r[i]} and {r[j]}')
    return flag  
```

```python
# diagonal atack
def diagonalattack(r, c, N):
    flag = False
    print('\ndiagonal attack: (if any)')
    for i in range(0,N):
        for j in range(i+1,N):
            if(r[i]-r[j] == c[i]-c[j]):
                flag = True
                print(f'row:{r[i]} column:{c[i]} and row:{r[j]} column:{c[j]}')
    return flag
```

```python
# implement a method to print a state is valid or not
def isValid(r, c, N):
    
    row = rowattack(r, N) 
    column = columnattack(c, N) 
    diagonal = diagonalattack(r, c, N)
    
    if row or column or diagonal:
        print("\n====State is INVALID====")
    else:
        print("\n====State is VALID====")
```

```python
def showBoard(r, c, N):
    print('r =', r)
    print('c =', c)

    board = ""
    newboard = ""
    for i in range(0,N*N): board += "*"

    for i in range(0, len(r)): 
        index = (r[i]-1) * N + c[i] - 1
        board = board[:index] + "Q" + board[index+1:]

    for i in range(0, len(board)): 
        if i%N == 0: newboard += "\n" + board[i] + " "
        else: newboard += board[i] + " "

    print(newboard)
```

```python
# input row and col (two lists)
N = 4 
r = []
c = []

for i in range(0, N):
    r1 = random.choice(range(1, N+1))
    c1 = random.choice(range(1, N+1))
    
   
    while(True):
        for i in range(0,len(r)):
            for j in range(i+1,len(c)):
                if(r[i] == r1 and c[i] == c1):
                    print(f'repetition : row:{r[i]}, column:{c[i]}')
                    r1 = random.choice(range(1, N+1))
                    c1 = random.choice(range(1, N+1))
                    continue
        break
    r.append(r1)
    c.append(c1)

showBoard(r, c, N)
isValid(r, c, N)
```
